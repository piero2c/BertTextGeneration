Carrega bertimbau

````python
# Carrega a implementação em python do tokenizador, no meu caso, escolhi isso para diminuir o pico de memória
# no processo de tokenização
tokenizer = AutoTokenizer.from_pretrained('neuralmind/bert-base-portuguese-cased', use_fast=True)
````

````python
tokenizer.add_special_tokens({
    'additional_special_tokens': ['[unused1]']
})
````

````python
_CLS, _MASK, _EOS, _SEP  = tokenizer.encode('[MASK] [unused1]')
````

Cria tf.dataset

````python
def load_first(fpath, n=1_000):
    with open(fpath, encoding='utf-8') as f_iter:
        sample = [next(f_iter).rstrip() for _ in range(n)]
    return sample
````

````python
reviews = load_first('../prepared_data/reviews.txt', n=1_000)
titles = load_first('../prepared_data/titles.txt', n=1_000)
labels = [int(label) for label in load_first('../prepared_data/labels.txt', n=1_000)]
````

````python
labels[0:10]
````

Aplica tokenizador (isso pode levar alguns minutos e irá consumir uma quantidade significativa de memória)

````python
dataset = tokenizer(reviews, text_pair=titles, padding=True,
                    max_length=58 + 11 + 2,
                    truncation=True)
````

````python
tokenizer.decode(dataset['input_ids'][35])
````

````python
tokenizer.decode(dataset['input_ids'][36])
````

Converte labels do formato 
```
[token_id1, token_id2, ...]
```

para o formato one-hot

````python
%%time
train_dataset = tf.data.Dataset.from_tensor_slices((
    dict(dataset),
    np.array(labels, dtype='int32')
))
````

## Modelo

````python
training_args = TFTrainingArguments(
    output_dir='./results', 
    max_steps=100_000,
    per_device_train_batch_size=1,  # batch size per device during training
    warmup_steps=500,                # number of warmup steps for learning rate scheduler
    logging_dir='./logs',            # directory for storing logs
    logging_first_step=True,
    logging_steps=1
)
````

Carrega BERT pré-treinado para o português

````python
with training_args.strategy.scope():
    bert = TFAutoModelForSequenceClassification.from_pretrained('neuralmind/bert-base-portuguese-cased',
                                                                num_labels=tokenizer.vocab_size,
                                                                from_pt=True)
````

Treinador

````python
trainer = TFTrainer(
    model=bert,                         # the instantiated 🤗 Transformers model to be trained
    args=training_args,                  # training arguments, defined above
    train_dataset=train_dataset,    # tensorflow_datasets training dataset
)
````

````python
for batch in train_dataset.batch(1):
    break
````

------

````python
trainer.train()
````

Define otimizador

````python
optimizer = tf.keras.optimizers.Adam(lr=1e-5)
````

````python
for i, (batch_input, batch_labels) in enumerate(train_dataset.batch(1)):
    with tf.GradientTape() as tape:
        output = bert(batch_input, labels=batch_labels)
        loss_value = output[0]

    grads = tape.gradient(loss_value, bert.trainable_weights)
    optimizer.apply_gradients(zip(grads, bert.trainable_weights))
    print(f'Iter {i}: {loss_value}')
````
